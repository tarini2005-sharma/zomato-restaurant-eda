```python
import pandas as pd
import seaborn as sns 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("Zomato-data-.csv")
df.head(5)
```

```python
df.info()
df.shape
```

```python
missing_values=df.isnull().sum()
missing_values
```

```python
print(df.isnull().any())
```

```python
df['online_order']=df['online_order'].astype('category')
```

```python
df['listed_in(type)']=df['listed_in(type)'].astype('category')
```

```python
print(df.dtypes)
```

```python
df.describe()
```

**Q1:What is the average rating of restaurants?**

```python
df['rate']=pd.to_numeric(df['rate'],errors='coerce')
```

```python
df['rate'].dtypes
```

```python
df['rate'].unique()
round(df['rate'].mean(),2)
```

**After cleaning the rating column and converting it into numeric format, the average rating of restaurants in the dataset was found to be 3.63.**

**This indicates that most restaurants listed on Zomato are moderately well-rated, reflecting satisfactory food quality and service.**

**The average rating falls within the typical Zomato rating range (3.5–4.0), suggesting a balanced mix of good and average restaurants.**

**This insight highlights that users generally have a positive dining experience across the majority of restaurants in the dataset**

**Q2:Which restaurants have the highest ratings?**


```python
df.loc[df['rate'].idxmax()]
```

**The highest-rated restaurant in the dataset is ONESTA which has a rating of 4.6, reflecting outstanding customer reviews and service quality.**

**Q3:Which cities have the highest average restaurant ratings?**

```python
avg_rating=df['rate'].mean()
avg_rating
```

```python
above_avg=df.loc[df['rate']>avg_rating]
below_avg=df.loc[df['rate']<avg_rating]
above_avg
below_avg
```

**- No restaurant has a rating exactly equal to the average (3.63).**
**- Restaurants with ratings close to the average include ROSEWOOD INTERNATIONAL HOTEL - BAR & RESTAURANT, CAFE COFFEE DAY, and MY TEA HOUSE, all rated 3.6.**  
**- These restaurants can be considered as representative of the average dining experience in the dataset.**  
**- Restaurants with ratings above 3.63 are highly rated, while those below 3.63 are rated lower than average.**

**Q4:What is the average cost for two people?**

```python
avg_cost=df['approx_cost(for two people)'].mean()
avg_cost
```

**The average cost for two people at restaurants in the dataset is approximately ₹ 418, giving an idea of the typical expense a customer can expect while dining out.**

**Q5:Is there a relationship between cost and rating?**

```python
plt.figure(figsize=(10,5))
sns.scatterplot(data=df, x='approx_cost(for two people)', y='rate')
plt.xlabel=('approx_cost(for two people)')
plt.ylabel=('rate')
plt.title=("relationship between cost and rate")
plt.show()



           
```

```python
corr=df['approx_cost(for two people)'].corr(df['rate'])
print("Correlation between cost and rate:",corr)
```

**The correlation between cost and rating is 0.275, indicating a weak positive relationship. This means higher costs tend to have slightly higher ratings, but the effect is not strong.**

**Q8:Are there cheap restaurants with high ratings?**


```python
min_cost=df['approx_cost(for two people)'].min()
min_cost

```

```python
x=df.loc[df['approx_cost(for two people)']<=min_cost]
x
```

**The analysis shows that the restaurants with the minimum cost for two people have ratings of 3.8 or lower, indicating that very cheap restaurants do not necessarily have high ratings in this dataset.**


**Q9:Are there expensive restaurants with low ratings?**

```python
max_cost=df['approx_cost(for two people)'].max()
max_cost
```

```python
y=df.loc[df['approx_cost(for two people)']>=max_cost]
y
```

**Yes, the dataset shows that some expensive restaurants have relatively low ratings (around 3.7), indicating that a higher price does not always guarantee better customer satisfaction.**
